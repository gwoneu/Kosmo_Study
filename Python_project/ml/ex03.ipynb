### 다항회귀

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/PolynomialRegressionData.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X, y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
```

```python
plt.figure(figsize=(10, 5), dpi=100)
plt.scatter(X, y, color='#C3D5F9')
plt.plot(X, reg.predict(X), color='#9EB5E4')
plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.score(X, y)
```

```python
#데이터 읽기
df = pd.read_csv('./data/PolynomialRegressionData.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
#X독립변수를 다항으로 변경
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
```

```python
#선형객체 만들어서 학습
reg = LinearRegression()
reg.fit(X_poly, y)
```

```python
#X 범위를 0.1단위로 생성
import numpy as np
X_range = np.arange(min(X), max(X), 0.1)
X_range = X_range.reshape(-1, 1)
```

```python
#그래프출력
plt.figure(figsize=(10, 5), dpi=50)
plt.scatter(X, y, color='blue')
plt.plot(X_range, reg.predict(poly_reg.fit_transform(X_range)), color='green')
plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
#예측
pred = reg.predict(poly_reg.fit_transform([[4.5]]))
pred
```
